```python
# 1. import libraries
import requests # to download html code
from bs4 import BeautifulSoup # to navigate through the html code
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.billboard.com/charts/hot-100/"
```

```python
# 3. download html with a get request. Use the function request.get() and store the output in response
response = requests.get(url)
# 200 status code means OK! response.status_code
print(response.status_code)
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.text, 'html.parser')
# 4.2. check that the html code looks like it should
#print(soup.prettify())
```

```python
#lets try to get the 1st title

soup.select('li>h3')[0].get_text().strip()
```

```python
#lets have a look at this class

soup.select('li>span')
```

```python
#lets get the 1st artist

soup.select('li > span.c-label.a-no-trucate')[0].get_text().strip()
```

```python
#time to create a list of the top 100 songs

top_100_songs=[]

for elem in range(100):
    top_100_songs.append(soup.select('li>h3')[elem].get_text().strip())
```

```python
#same thing for the artists

top_100_artists=[]

for elem in range(100):
    top_100_artists.append(soup.select('li > span.c-label.a-no-trucate')[elem].get_text().strip())
```

```python
#lets create the dataframe with artist and their respective song

top_100_df = pd.DataFrame({'artist': top_100_artists, 'song': top_100_songs})
```

```python
top_100_df.to_csv('top_100_scraped.csv',index=False)
```
